# 游늷 Extracci칩n(E - Extract)

```python
import requests
import pandas as pd

url = "https://raw.githubusercontent.com/sthemonica/alura-voz/refs/heads/main/Dados/Telco-Customer-Churn.json"

response = requests.get(url)
data = response.json()

df = pd.DataFrame(data)

df
```

```python
# Columnas con diccionarios
columnas_dict = ['customer', 'phone', 'internet', 'account']
```

```python
for col in columnas_dict:
    df_temp = pd.json_normalize(df[col])
    df = pd.concat([df, df_temp], axis=1)
    df.drop(columns=[col], inplace=True)
```

```python


df_plano=df
df_plano
```

```python
# Agrupar por tenure y Churn
df_grouped = df_plano.groupby(['tenure', 'Churn']).size().unstack(fill_value=0)

# Calcular el % de churn por tenure
df_grouped['churn_rate'] = (df_grouped.get('Yes', 0) / df_grouped.sum(axis=1)) * 100

# Mostrar el resultado
print(df_grouped[['churn_rate']])
```

```python
fig = px.line(df_grouped.reset_index(),
              x='tenure',
              y='churn_rate',
              markers=True,
              title='Tasa de Churn (%) por Tenure')
fig.show()
```
